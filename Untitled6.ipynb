```python
# Load embeddings (replace 'your_embeddings.npy' with your actual file)
embeddings = np.load("task_embeddings_new.npy")

# Check shape of data
print("Embedding Shape:", embeddings.shape)  # Example: (31747, 768)
```

```python
# Apply UMAP for 15D reduction
umap_model = umap.UMAP(n_components=15, n_neighbors=30, min_dist=0.1, metric='cosine', random_state=42)
umap_embeddings = umap_model.fit_transform(embeddings)

# Apply PCA for 15D reduction
pca_model = PCA(n_components=15)
pca_embeddings = pca_model.fit_transform(embeddings)

print("UMAP Reduced Shape:", umap_embeddings.shape)
print("PCA Reduced Shape:", pca_embeddings.shape)
```

```python
# Define number of clusters
num_clusters = 250

# Apply KMeans
kmeans = KMeans(n_clusters=num_clusters, random_state=42, n_init=10)
kmeans_labels = kmeans.fit_predict(umap_embeddings)

# Apply Spectral Clustering
spectral = SpectralClustering(n_clusters=num_clusters, affinity='nearest_neighbors', random_state=42)
spectral_labels = spectral.fit_predict(umap_embeddings)

# Apply Gaussian Mixture Model (GMM)
gmm = GaussianMixture(n_components=num_clusters, random_state=42)
gmm_labels = gmm.fit_predict(umap_embeddings)
```

```python
# Function to compute clustering metrics
def evaluate_clustering(embeddings, labels, method_name):
    sil_score = silhouette_score(embeddings, labels)
    ch_score = calinski_harabasz_score(embeddings, labels)
    db_score = davies_bouldin_score(embeddings, labels)

    print(f"\n📊 {method_name} Clustering Performance:")
    print(f"Silhouette Score: {sil_score:.4f}")
    print(f"Calinski-Harabasz Score: {ch_score:.4f}")
    print(f"Davies-Bouldin Score: {db_score:.4f}")
    return sil_score, ch_score, db_score

# Evaluate each method
kmeans_scores = evaluate_clustering(umap_embeddings, kmeans_labels, "KMeans")
spectral_scores = evaluate_clustering(umap_embeddings, spectral_labels, "Spectral Clustering")
gmm_scores = evaluate_clustering(umap_embeddings, gmm_labels, "GMM")
```

```python
# Apply UMAP for 10D reduction (optimized parameters)
umap_model_optimized = umap.UMAP(n_components=10, n_neighbors=15, min_dist=0.05, metric='cosine', random_state=42)
umap_embeddings_optimized = umap_model_optimized.fit_transform(embeddings)

print("Optimized UMAP Shape:", umap_embeddings_optimized.shape)
```

```python
from hdbscan import HDBSCAN

# Apply HDBSCAN (detects number of clusters automatically)
hdbscan_model = HDBSCAN(min_cluster_size=20, min_samples=10, metric='euclidean')
hdbscan_labels = hdbscan_model.fit_predict(umap_embeddings_optimized)

# Evaluate HDBSCAN
evaluate_clustering(umap_embeddings_optimized, hdbscan_labels, "HDBSCAN")
```

```python
from sklearn.cluster import MiniBatchKMeans

# Apply KMeans with 300 clusters
kmeans_optimized = KMeans(n_clusters=300, random_state=42, n_init=10)
kmeans_optimized_labels = kmeans_optimized.fit_predict(umap_embeddings_optimized)

# Apply MiniBatch KMeans (faster)
mini_kmeans = MiniBatchKMeans(n_clusters=300, batch_size=500, random_state=42)
mini_kmeans_labels = mini_kmeans.fit_predict(umap_embeddings_optimized)

# Evaluate KMeans
evaluate_clustering(umap_embeddings_optimized, kmeans_optimized_labels, "Optimized KMeans")
evaluate_clustering(umap_embeddings_optimized, mini_kmeans_labels, "MiniBatch KMeans")
```

```python
# Load embeddings (.npy file)
embeddings_path = "/content/task_embeddings_new.npy"
embeddings = np.load(embeddings_path)

# Load clustered dataset (.xlsx file) - Ensure 'Cluster' column exists
df = pd.read_excel("Clustered_Dataset.xlsx")

# Check data structure
print("Embeddings shape:", embeddings.shape)
print(df.head())
```

```python
# Load original dataset
df_original = pd.read_excel("/content/actions_count.xlsx")  # 5554 rows
df_filtered = pd.read_excel("/content/Clustered_Dataset.xlsx")  # 5315 rows (filtered)

# Find the index mapping (assuming df_filtered retains original index)
valid_indices = df_filtered.index  # Indices of kept rows

# Filter original embeddings using valid indices
filtered_embeddings = embeddings[valid_indices]

# Confirm matching shape
print("Filtered Embeddings Shape:", filtered_embeddings.shape)
print("Filtered DataFrame Shape:", df_filtered.shape)
```

```python
import umap

# Reduce embeddings to 2D
umap_2d = umap.UMAP(n_components=2, random_state=42)
embeddings_2d = umap_2d.fit_transform(filtered_embeddings)

# Store results in the filtered DataFrame
df_filtered["UMAP_1"] = embeddings_2d[:, 0]
df_filtered["UMAP_2"] = embeddings_2d[:, 1]

# Reduce embeddings to 3D
umap_3d = umap.UMAP(n_components=3, random_state=42)
embeddings_3d = umap_3d.fit_transform(filtered_embeddings)

df_filtered["UMAP_3D_1"] = embeddings_3d[:, 0]
df_filtered["UMAP_3D_2"] = embeddings_3d[:, 1]
df_filtered["UMAP_3D_3"] = embeddings_3d[:, 2]
```

```python
import umap
import pandas as pd
import numpy as np

# Load your latest embeddings (Filtered embeddings)
filtered_embeddings_path = "/content/task_embeddings_new.npy"  # Update path if needed
filtered_embeddings = np.load(filtered_embeddings_path)

# Load your clustered dataset (Ensure it has cluster labels)
df = pd.read_excel("/content/clustered_tasks_with_umap.xlsx")

# Check if rows match
if len(df) != len(filtered_embeddings):
    print(f"⚠️ Mismatch detected: DataFrame rows = {len(df)}, Embeddings = {len(filtered_embeddings)}")
else:
    print("✅ Rows match correctly!")

# Apply UMAP for 2D visualization
umap_2d = umap.UMAP(n_components=2, random_state=42)
umap_2d_embeddings = umap_2d.fit_transform(filtered_embeddings)

# Store UMAP values in DataFrame
df["UMAP_1"] = umap_2d_embeddings[:, 0]
df["UMAP_2"] = umap_2d_embeddings[:, 1]

# Save updated file (Optional)
df.to_excel("/content/clustered_tasks_with_umap.xlsx", index=False)
```

```python
import plotly.express as px

fig = px.scatter(df, x="UMAP_1", y="UMAP_2", color=df["Cluster"].astype(str), hover_data=["tag_action"])

# Reduce marker size
fig.update_traces(marker=dict(size=5))  # Adjust 'size' as needed

fig.update_layout(title="Interactive Scatter Plot of Topics", width=800, height=600)
fig.show()
```

```python
import umap
import pandas as pd
import numpy as np

# Load your filtered embeddings (Make sure these match your clustered dataset)
filtered_embeddings_path = "task_embeddings_new.npy"
filtered_embeddings = np.load(filtered_embeddings_path)

# Load the dataset that already has UMAP_1, UMAP_2, and Cluster labels
df = pd.read_excel("/content/clustered_tasks_with_umap.xlsx")

# Check if row counts match
if len(df) != len(filtered_embeddings):
    print(f"⚠️ Mismatch detected: DataFrame rows = {len(df)}, Embeddings = {len(filtered_embeddings)}")
else:
    print("✅ Rows match correctly!")

# Apply UMAP for 3D visualization
umap_3d = umap.UMAP(n_components=3, random_state=42)
umap_3d_embeddings = umap_3d.fit_transform(filtered_embeddings)

# Store UMAP 3D values in DataFrame
df["UMAP_3D_1"] = umap_3d_embeddings[:, 0]
df["UMAP_3D_2"] = umap_3d_embeddings[:, 1]
df["UMAP_3D_3"] = umap_3d_embeddings[:, 2]

# Save the updated dataset with 3D UMAP values
df.to_excel("/content/clustered_tasks_with_umap_3d.xlsx", index=False)

print("✅ UMAP 3D values added and dataset saved successfully!")
```

```python
import plotly.express as px

# Create 3D scatter plot
fig = px.scatter_3d(
    df,
    x="UMAP_3D_1",
    y="UMAP_3D_2",
    z="UMAP_3D_3",
    color=df["Cluster"].astype(str),
    opacity=0.8
)

# Reduce marker size
fig.update_traces(marker=dict(size=3))  # Adjust size as needed

# Update layout
fig.update_layout(title="3D Cluster Visualization (UMAP)", height=700)

# Show figure
fig.show()
```
